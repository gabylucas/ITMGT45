```python
#PROBLEM 2: TIC-TAC-TOE BOARD

board1 = [
['X','X','O'],
['O','X','O'],
['O','','X'],
]

board2 = [
['X','X','O'],
['O','X','O'],
['','O','X'],
]

board3 = [
['O','X','O'],
['','O','X'],
['X','X','O'],
]

board4 = [
['X','X','X'],
['O','X','O'],
['O','','O'],
]

board5 = [
['X','X','O'],
['O','X','O'],
['X','','O'],
]

board6 = [
['X','X','O'],
['O','X','O'],
['X','',''],
]

def get_winner(board):
    winner = None
    
   ## check diagonals
    x = [board[g][g] for g,v in enumerate(board)]
    if all(num == x[0] for num in x):
        winner = x[0]
    y = [board[g][2-g] for g,v in enumerate(board)]
    if all(num == y[0] for num in y):
        winner = y[0]
        
        
    ## check rows
    c = 0
    while c < 3:
        z = board[c]
        if all(num == z[0] for num in z):
            winner = z[0]   
        c+=1
        
      ## check columns
    n=0
    while n<3:
        a = [column[n] for column in board]
        if all(num == a[0] for num in a):
            winner = a[0]
        n+=1
        
    return winner

print(get_winner(board1))
```

```python
#PROBLEM 3: ESTIMATED TIME OF ARRIVAL

locations = {
    "admu":{
        "name":"Ateneo de Manila University",
        "latitude":14.6394557,
        "longitude":121.0758975        
    },
    "upd":{
        "name":"University of the Philippines-Diliman Campus",
        "latitude":14.6538519,
        "longitude":121.0663348
    },
    "dlsu":{
        "name":"De La Salle University-Taft",
        "latitude":14.5647346,
        "longitude":120.9909876

    }, # <-- Note that Python does not complain about this extra comma. This is perfectly okay.
       #     The creators of Python saw that dictionaries tend to be changed a lot, so allowing this is a
       #     convenience for developers who need to add elements when needed.
}

legs = {
     ("upd","admu"):{
         "travel_time_mins":10
     },
     ("admu","dlsu"):{
         "travel_time_mins":35
     },
     ("dlsu","upd"):{
         "travel_time_mins":55
     }
}

    
source=input("From: ")
destination=input("To: ")

def eta(source, destination): 
    upd_admu = legs.get(("upd","admu"))["travel_time_mins"]
    admu_dlsu = legs.get(("admu","dlsu"))["travel_time_mins"]
    dlsu_upd = legs.get(("dlsu","upd"))["travel_time_mins"]
    if "upd" in source and "admu" in destination:
        x = upd_admu
    elif "upd" in source and "dlsu" in destination:
        x = upd_admu + admu_dlsu
    elif "admu" in source and "dlsu" in destination:
        x = admu_dlsu
    elif "admu" in source and "upd" in destination:
        x = admu_dlsu + dlsu_upd
    elif "dlsu" in source and "upd" in destination:
        x = dlsu_upd
    elif "dlsu" in source and "admu" in destination:
        x = dlsu_upd + upd_admu
    else:
        x = "Error"
    return x

print("Your ETA is: ", eta(source, destination))
```
